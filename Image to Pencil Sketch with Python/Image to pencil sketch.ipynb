#### 1.1 Importing the required modules

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#### 1.2 Reading the image

```python
image = cv2.imread("image.jpg")
while True:
    cv2.imshow("Input Image--Press ESC to exit", image)
    if cv2.waitKey(2) == 27:
      break
# Window shown waits for ESC key pressing event
cv2.destroyAllWindows()
```

#### 1.3 Analyzing the image

```python
image.shape
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

pixels = np.array(image)

plt.imshow(pixels)
plt.show()
```

#### 1.4 Grey Scaling the image

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

#Showing the gray scaled image
cv2.imshow('Grayscale', gray_img)
cv2.waitKey(0) 
 
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```

#### 1.5 Inverting the gray scale image

```python
inverted_img = 255 - gray_img
cv2.imshow("Inverted", inverted_img)
cv2.waitKey(0)
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```

#### 1.6 Bluring the image

```python
blurred = cv2.GaussianBlur(inverted_img, (613, 613), 0)
cv2.imshow("Blurred", blurred)
cv2.waitKey(0)
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```

#### 1.7 Final Result-- Pencil Sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```
